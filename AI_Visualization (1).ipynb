```
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```
Algorithm = ['CNN-Adam Optimizer', 'K-Nearest Neighbors', 'CNN-SGD Optimizer', 'Multi-Layer Perceptron', 'Support Vector Machine - rbf', 'Support Vector Machine - Linear', 
         'Logistic Regression', 'Random Forest Classifier', 'Squeezenet/Vgg16']
Time = [7.5, 40.5, 8.3, 59.5, 89, 210, 2, 3.3, 64]
Accuracy = [93.4, 86.6, 85.8, 84.7, 82, 78.1, 75.6, 70.1, 22.5]
data = {'Algorithm': Algorithm, 'Time taken (in mins)': Time, 'Accuracy (in %)': Accuracy}
df = pd.DataFrame(data)
df[:9]
```

```
plt.title('Time taken')
sns.barplot(x='Algorithm', y='Time taken (in mins)', data=df, palette='pastel')
```

```
plt.title('Accuracy')
sns.lineplot(x='Algorithm', y='Accuracy (in %)', data=df, sort=False)
```

```
#Create combo chart
fig, ax1 = plt.subplots(figsize=(10,6))
color = 'tab:green'
#bar plot creation
ax1.set_title('Results', fontsize=16)
ax1.set_xlabel('Algorithm', fontsize=16)
ax1.set_ylabel('Time taken (in mins)', fontsize=16)
ax1 = sns.barplot(x='Algorithm', y='Time taken (in mins)', data = df, palette='pastel')
ax1.tick_params(axis='y')
#specify we want to share the same x-axis
ax2 = ax1.twinx()
color = 'tab:red'
#line plot creation
ax2.set_ylabel('Accuracy (%)', fontsize=16)
ax2 = sns.lineplot(x='Algorithm', y='Accuracy (in %)', data = df, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
#show plot
plt.show()
```
